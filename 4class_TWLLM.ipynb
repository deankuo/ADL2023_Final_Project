```python
df = pd.read_excel("/content/instruction_testing.xlsx")
```

```python
df = df.drop_duplicates()
```

```python
df['information'] = df['information'].map({1: '是', 0: '否'})
df['justification'] = df['justification'].map({1: '是', 0: '否'})
df['change'] = df['change'].map({1: '是', 0: '否'})
df['sanctions'] = df['sanctions'].map({1: '是', 0: '否'})
```

```python
def information_trans(df):
  df['instruction'] = ''
  df['output'] = ''
  for i in range(len(df)):
    text = df.loc[i, 'text']
    instruction = f'''一個要求資訊的質詢包含了對政策客觀數字的提問或是提出要求書面資料和報告。
            請分析下列議員對政府官員的質詢是否有包含"要求資訊"。
            質詢:「{text}」
            選項: -是 -否'''
    df.loc[i, 'instruction'] = instruction
    df['output'][i] = df['information'][i]

def justification_trans(df):
  df['instruction'] = ''
  df['output'] = ''
  for i in range(len(df)):
    text = df['text'][i]
    instruction = f'''一個要求說明的質詢包含了尋求官員的主觀看法或是意見或是要求官員對政策進行說明。
            請分析下列議員對政府官員的質詢是否有包含"要求說明"。
            質詢:「{text}」
            選項: -是 -否'''
    df['instruction'][i] = instruction
    df['output'][i] = df['justification'][i]

def change_trans(df):
  df['instruction'] = ''
  df['output'] = ''
  for i in range(len(df)):
    text = df['text'][i]
    instruction = f'''一個要求改變的質詢包含要求政府對既有政策做出改變、移除或新增政策、或是要求政府做出具體行動和承諾。
            請分析下列議員對政府官員的質詢是否有包含"要求改變"。
            質詢:「{text}」
            選項: -是 -否'''
    df['instruction'][i] = instruction
    df['output'][i] = df['change'][i]

def sanctions_trans(df):
  df['instruction'] = ''
  df['output'] = ''
  for i in range(len(df)):
    text = df['text'][i]
    instruction = f'''一個威脅譴責的質詢包含了批評指責或辱罵政府官員和機構、向政府官員和機構威脅提出政風司法調查、或是要求政府官員下台和承擔政治責任。
            請分析下列議員對政府官員的質詢是否有包含"威脅譴責"。
            質詢:「{text}」
            選項: -是 -否'''
    df['instruction'][i] = instruction
    df['output'][i] = df['sanctions'][i]
```

```python
def create_sample(df):

    # Create four identical copies of the sampled DataFrame
    task1_df = df.copy()
    task2_df = df.copy()
    task3_df = df.copy()
    task4_df = df.copy()

    # Apply each transformation function to each copy
    information_trans(task1_df)
    justification_trans(task2_df)
    change_trans(task3_df)
    sanctions_trans(task4_df)

    # Concatenate the four DataFrames into a single DataFrame
    result_df = pd.concat([task1_df, task2_df, task3_df, task4_df], ignore_index=True)
    result_df = result_df[['instruction', 'output']]

    return result_df
```

```python
result_df = create_sample(df)
```
